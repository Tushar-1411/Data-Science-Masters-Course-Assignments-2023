```python
# Question 1 

# read_csv(): This function is used to read data from a CSV file and return a DataFrame object.

df = pd.read_csv("data.csv")
```

```python
# groupby(): This function is used to group data in a DataFrame by one or more columns, and then perform aggregate functions on each group.

df = pd.read_csv("data.csv")
grouped = df.groupby("Category")
```

```python
# merge(): This function is used to combine two or more DataFrames into a single DataFrame.

df1 = pd.read_csv("data1.csv")
df2 = pd.read_csv("data2.csv")
merged = pd.merge(df1, df2, on="key")
```

```python
# pivot_table(): This function is used to create a pivot table from a DataFrame.

df = pd.read_csv("data.csv")
pivot = df.pivot_table(index="Category", values="Sales", aggfunc="sum")
```

```python
# Question 2

def reindex_df(df):
    new_index = pd.RangeIndex(start=1, stop=len(df)*2, step=2)
    return df.reset_index(drop=True).set_index(new_index)
```

```python
# Question 5

""" Both DataFrame.size and DataFrame.shape are used to get information about the dimensions of a Pandas DataFrame, but they have different meanings:

1. DataFrame.size returns the total number of elements in the DataFrame, which is equal to the number of rows multiplied by the number of columns.

2. DataFrame.shape returns a tuple that contains the number of rows and the number of columns in the DataFrame, in that order.

Here's an example to illustrate the difference: 
"""
# Create a 2x3 DataFrame
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4], 'C': [5, 6]})

# Print the size and shape of the DataFrame
print('Size:', df.size)
print('Shape:', df.shape)
```

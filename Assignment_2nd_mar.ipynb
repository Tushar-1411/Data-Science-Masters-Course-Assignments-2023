### Question 1
Matplotlib is used for data visualization tasks, such as creating line plots, scatter plots, bar plots, histograms, and more.

The Pyplot module of Matplotlib provides a simple interface for creating a wide range of plots. Some of the plots that can be created using Pyplot module are:

1. Line Plot
2. Scatter Plot
3. Bar Plot
4. Histogram
5. Box Plot

```python
# Question 2
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("Normal_X")
plt.ylabel("Normal_Y")
plt.title("Random Normal Distribution X vs Y")
```

```python
# Question 3

"""The subplot() function is used in Matplotlib to create a grid of subplots within a single figure. 
It is used when we want to create multiple plots within a single figure."""

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(221)
plt.plot(x1,y1)
plt.subplot(222)
plt.plot(x2,y2)
plt.subplot(223)
plt.plot(x3,y3)
plt.subplot(224)
plt.plot(x4,y4)
```

```python
# Question 4

""" Bar plots are useful when you want to visualize and compare the quantities or frequency of different categories or groups of data. 
They are commonly used in business, finance, economics, and social sciences to display data in a clear and concise way. """

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel("Company"); plt.ylabel("Profit"); plt.title("Company Profits")
```

```python
# Question 5

""" A box plot is a type of graphical representation that provides a summary of the distribution of a dataset. 
It is also called a box-and-whisker plot.
Box plots are used to visually represent the distribution of data, including the center, spread, and skewness of the data. 
They can be used to compare different groups or categories of data and identify potential outliers."""

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig, ax = plt.subplots()
ax.boxplot([box1,box2])
plt.show()
```

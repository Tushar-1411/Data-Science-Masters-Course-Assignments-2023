```python
# Question 1

import scipy.stats as stats

sample_mean = 50
sample_std_dev = 5
sample_size = 100  # Assuming a sample size of 100 for demonstration purposes
confidence_level = 0.95

# Calculate the standard error of the mean
SE = sample_std_dev / (sample_size ** 0.5)

# Find the critical z-value
critical_z = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the confidence interval
lower_bound = sample_mean - critical_z * SE
upper_bound = sample_mean + critical_z * SE

print(f"The 95% confidence interval for the sample mean is ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# Question 2

import numpy as np
from scipy.stats import chisquare

# Observed counts of colors in the bag
observed_counts = np.array([30, 22, 24, 5, 2, 17])  # Blue, Orange, Green, Yellow, Red, Brown

# Expected counts based on the given distribution
expected_counts = np.array([20, 20, 20, 10, 10, 20])

# Perform the chi-square test
chi2, p_value = chisquare(observed_counts, f_exp=expected_counts)

# Set the significance level (alpha)
alpha = 0.05

# Make a decision based on the p-value
if p_value < alpha:
    print("The distribution of M&M colors in the bag significantly differs from the expected distribution.")
else:
    print("There is no significant evidence that the distribution of M&M colors in the bag differs from the expected distribution.")
```

```python
# Question 3

import numpy as np
from scipy.stats import chi2_contingency

# Given data in the form of a contingency table
observed_data = np.array([[20, 10, 15],
                          [15, 25, 20]])

# Perform the chi-square test
chi2, p_value, dof, expected = chi2_contingency(observed_data)

print("Chi-square Statistic:", chi2)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
# Question 4

import scipy.stats as stats

sample_size = 500
smokers_count = 60
confidence_level = 0.95

# Calculate the sample proportion
sample_proportion = smokers_count / sample_size

# Calculate the standard error of the proportion
SE = (sample_proportion * (1 - sample_proportion) / sample_size) ** 0.5

# Find the critical z-value
critical_z = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the confidence interval
lower_bound = sample_proportion - critical_z * SE
upper_bound = sample_proportion + critical_z * SE

print(f"The 95% confidence interval for the proportion of individuals who smoke is "
      f"({lower_bound:.4f}, {upper_bound:.4f})")
```

```python
# Question 6

import numpy as np
import matplotlib.pyplot as plt

# Define the degrees of freedom
degrees_of_freedom = 10

# Generate x values for the chi-square distribution
x = np.linspace(0, 30, 500)

# Calculate the chi-square probability density function
pdf = stats.chi2.pdf(x, df=degrees_of_freedom)

# Plot the chi-square distribution
plt.plot(x, pdf, label='Chi-square (df=10)')
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.title('Chi-square Distribution')
plt.axvspan(15, 30, color='gray', alpha=0.3, label='Chi-square Statistic = 15')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question 8

import numpy as np
from scipy.stats import chisquare

# Observed counts of tails and heads
observed_counts = np.array([45, 55])

# Expected counts for a fair coin (50% probability for each outcome)
expected_counts = np.array([50, 50])

# Perform the chi-square test
chi2, p_value = chisquare(observed_counts, f_exp=expected_counts)

# Set the significance level (alpha)
alpha = 0.05

# Make a decision based on the p-value
if p_value < alpha:
    print("The observed frequencies significantly differ from the expected frequencies of a fair coin.")
else:
    print("There is no significant evidence that the observed frequencies differ from the expected frequencies of a fair coin.")
```

```python
# Question 10

import numpy as np
from scipy.stats import chi2_contingency

# Given data in the form of a contingency table
observed_data = np.array([[200, 150, 150],
                           [225, 175, 100]])

# Perform the chi-square test
chi2, p_value, dof, expected = chi2_contingency(observed_data)

# Set the significance level (alpha)
alpha = 0.01

print("Chi-square Statistic:", chi2)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)

# Make a decision based on the p-value
if p_value < alpha:
    print("There is a significant association between chocolate preference and country of origin.")
else:
    print("There is no significant evidence of an association between chocolate preference and country of origin.")
```

```python
# Question 11

import scipy.stats as stats

sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean = 70
significance_level = 0.05

# Calculate the standard error of the mean
SE = sample_std_dev / (sample_size ** 0.5)

# Calculate the z-score
z_score = (sample_mean - population_mean) / SE

# Calculate the critical z-value for the chosen significance level
critical_z = stats.norm.ppf(1 - significance_level/2)

# Perform the two-tailed test
if abs(z_score) > critical_z:
    print("The population mean is significantly different from 70.")
else:
    print("There is no significant evidence to reject the hypothesis that the population mean is 70.")
```
